# U.S. Medical Insurance Costs

```python
import csv
import math
import matplotlib.pyplot as plt
```

```python
with open("insurance.csv") as insurance_csv:
    #defining a dictionary with data from the csv file
    insurance_dict = {}
    insurance_dict = csv.DictReader(insurance_csv)
    
    #Section for defining functions:
    
    def average_age(ages):
        total_years = 0
        for age in ages:
            total_years += float(age)
        av_age = math.floor(total_years/len(ages))
        print("The average age in this dataset is "+str(av_age)+" years.")
    
    
    #Section for defining variables:
    ages = []
    sex = {}
    region = {}
    num_of_children = {}
    smoker = {"smoker" : 0, "not a smoker" : 0}
    
    for element in insurance_dict: #filling all variables with data from the file
        ages.append(element["age"])
        if element["sex"] not in sex:
            sex[element["sex"]] = 1
        else:
            sex[element["sex"]] += 1
        if element["region"] not in region:
            region[element["region"]] = 1
        else:
            region[element["region"]] += 1
        if element["children"] not in num_of_children:
            num_of_children[element["children"]] = 1
        else:
            num_of_children[element["children"]] += 1
        if element["smoker"] == "yes":
            smoker["smoker"] += 1
        else:
            smoker["not a smoker"] += 1
        
    
    
    # Section for using the data to see some results

    average_age(ages)
    lists = sorted(num_of_children.items()) # sorted by key, return a list of tuples

    x, y = zip(*lists) # unpack a list of pairs into two tuples

    plt.plot(x, y)
    plt.xlabel("Number of children")
    plt.ylabel("Number of families")
    plt.title("Families with a certain amount of kids")
    plt.show()
    
    smoker_labels = 'smoker', 'no smoker'
    smoker_values = [smoker["smoker"], smoker["not a smoker"]]
    fig1, ax1 = plt.subplots()
    ax1.pie(smoker_values,labels=smoker_labels,autopct='%1.1f%%')
    ax1.axis('equal')
    plt.title("Smoker to non-smoker ratio")
    plt.show()
```
